# 二、网络层

网络层提供的是网际间的通信，提供了主机与主机之间的通信，需要注意的是：网络层提供的是不可靠的交付服务，其不保证通信质量，只是尽最大努力交付，因此没有建立连接释放连接这些操作。工作在网络层的协议主要有以下几种：

1. 地址解析协议ARP。

2. 逆地址解析协议RARP。

3. 网际控制报文协议ICMP。

4. 网际组管理协议IGMP。

## `IP`地址

>IP地址就是给因特网上的每一个主机（或路由器）的每一个接口分配一个在全世界范围内是唯一的32的标识符。

IP地址的编码分为三个阶段：

1. 分类的IP地址：这是最基本的编址方法，在1981年就通过了相应的标准协议。

2. 子网的划分：这是对最基本的编址方法的改进，其标准RFC950在1985年通过。

3. 构成超网：这是比较新的无分类编址方法，1993年提出后很快得到推广使用。

早期的IP地址采用分类编址的方法，一共将IP地址分为5类，其中常用的IP地址类别是A，B和C三类：

|网络类别|最大可以指派的网络数|第一个可指派的网络号|最后一个可指派的网络号|每个网络中最大主机数|
|:-:|:-:|:-:|:-:|:-:|

可以注意到：每个类别的网络号的最大编码网络数量都是在总数量上减1或2，这是由于网络号编码中全0的编码方案是不用的，而A类地址中127.0.0.1这个地址又被用做是本机的回环地址，因此A类地址的网络数减去2而其他类别的网络数减去的是1。在每种网络的可编码主机上依然是全0或者全1的编码方案不使用，因此主机编码方案也都要减去2。

## `ARP`协议

ARP协议的中文名称是地址解析协议，这是网络层中一个非常重要的协议，它的功能是将IP地址转化为物理地址。由于网络报文的传递最终还是要反映在底层的数据链路上的，因此将IP地址映射为物理地址就显得十分重要。一个十分现实的应用场景就是：假如A主机要与另一个网络中的主机进行通信，那么A主机将首先把网络报文交给A主机所在网络上的路由器（假定路由器的IP地址可知，至于为什么可知就需要了解路由表的建立过程），这就出现了一个问题：A主机如何知道路由器的物理地址呢？如果不知道路由器的物理地址则无法进行通信。解决方案就是使用ARP协议。

ARP协议的工作过过程十分简单，由于一个网络内部可以理解为一个碰撞域，因此A主机可以利用数据链路层进行广播通信，当A主义需要运行ARP协议的时候就在本地碰撞域发送一个广播报文，报文的内容大概是：“我是xxx.xxx.xxx.xxx（其IP地址），硬件地址是：xx-xx-xx-xx-xx-xx-xx（物理地址），我想知道xxx.xxx.xxx.xxx（目标机器的IP地址）的硬件地址”。这个广播报文在碰撞域内部的所有主机都能收到，但只有IP地址符合的主机才响应这个报文，其余的主机将会丢弃这个报文。A主机接收到目标机器的响应报文后就会在其内部维护一个ARP高速缓存，将IP与硬件地址的映射关系保留下来，为了防止该缓存时间太久而过期（IP地址与物理地址的映射关系发生改变，或者网络拓扑改变），该缓存每过一段时间就会更新一次，ARP协议的运行过程如下图所示：



## `IP`数据报的报文格式

一个IP数据报由首部和数据两部分组成。首部的前一部分是固定长度，共20个字节，是所有IP数据报必须具有的。在首部的固定部分后面是一些可选字段，其长度时可变的：



在这里需要重点关注的是**标识**字段，**标志**以及**片偏移**字段，根据实际使用情况IP数据报的最大长度不用超过1500字节，最小长度不应超过576字节。如果超过1500字节就应该对其进行分片以保证IP报文的长度加上数据链路层首部后不超过数据链路层报文的最大长度。如果需要对IP报文进行分片那么首先需要产生一个16位的标识，注意该标识并不是TCP中的序号，它仅仅用来标识一个IP数据报，也就是说如果一个IP数据报因为太长而被分片，那么每个分片都会带有一个相同的标识来告诉IP层：这些分片同属于同一个IP数据报。接下来要面对的问题就是如何将分片好的IP报文组装回去，在这里就需要用到标志字段以及片偏移字段。片偏移字段表明该分片在整个IP报文中的位置，而标志字段则告诉我们该分片后面是否还有分片以及该分片是不是最后一个分片等信息。

## 划分子网

通常情况下一个A类网和B类网的主机数目是非常大的，而一个单位或者组织往往用不了这么多的主机数目这可能就会造成IP地址的浪费，这就是传统的两级IP地址或者说分类IP地址的缺点。解决这一问题的思路就是**划分子网**。一个能支持很多主机的网络可以在其内部进行子网划分，这样一个大的网络就能够被划分为很多子网，再将这些子网分配给不同的组织或机构。采用子网划分的IP编址方案被称作是三级IP地址：

$$IP地址::=\{<网络号>,<子网号>,<主机号>\}$$

在划分子网的方案中利用到子网掩码进行子网划分，子网掩码是由32位1和0组成的二进制串，它与IP地址进行与运算就能够得出目标IP的网络地址，再以此进行转发。子网掩码现在已经成为一个被广泛接受的网络标准，在网络节点的路由器内部的路由表中，不仅仅有网络号和下一跳地址，现在还包含子网掩码字段。由于不能仅仅从IP地址上看出其网络号，所以需要使用子网掩码，该子网掩码机器路由表的生成依然需要特定的协议去自适应生成。

## 构造超网

采用划分子网的形式虽然可以提高IP地址的利用率，但是这无疑会大大增加路由表的容量，原来一个B类网只在路由表中占用一个条目，但是现在可能由于子网划分一个B类网在路由表中要占据多得多的位置。为了解决这个问题就要用到构造超网技术。传统的划分子网方案依然是在分类IP地址编码的基础上对其进行子网的划分，这样的方案依然灵活度不够，在1993年正式提出了CIDR的RFC文档。现在CIDR已经成为因特网建议标准协议，CIDR的最主要特点有两个：

1. CIDR消除了传统的A类，B类和C类地址以及划分子网的概念，因而可以更加灵活充分地分配网络地址，CIDR采用变长网络前缀的解决方案：将32位的IP地址划分为两个部分。前面的部分是“网络前缀”用以指明网络，后面的部分则用来指明主机。因此CIDR将网络地址又从划分子网的三级编址方案变回了传统的两级编址方案：

$$IP地址::=\{<网络前缀>,<主机号>\}$$

2. CIDR将网络前缀都相同的连续的IP地质组成一个“CIDR地址块”。我们只要知道CIDR地址块中的任何一个地址就可以知道这个地址块的起始地址和最大地址，以及地址块中的地址数目。

之所以用构造超网来指代CIDR不是因为CIDR可以将多个不同的网络够造成同一个网络（当然实际操作中肯定可以这么干，这里仅仅是从概念上），而是说一个CIDR地址块相当于多个网络。例如网络前缀为13的地址块能容纳512K个主机，这相当于8个B类网或者2048个C类网的主机容量，超网的概念是从这个意义上来说的。由于构造超网比较难以理解，因此下面用一个例子来解释：

>有一个ISP（互联网服务提供商）拥有一个地址块206.0.64.0/18，现在有一所大学需要大约800个IP地址，向该ISP申请地址块。该大学内部又有4个系，分别为1系，2系，3系和4系。每个系的IP地址需求量分别为400,200,100和100，问该如何分配：

上面的问题非常典型，现在首先来分析：大学需要约800个IP地址，也就是说分配给它的地址块需要支持800台主机，由于二进制的关系只能给他分配一个可以容纳1024台主机的地址块（因为宁多勿少，如果分配512个主机的地址块明显不符合要求），已知$2^{10}=1024$，因此分配给大学的网络前缀要有$22(32-10)$位。用二进制的方式写出ISP拥有的这个地址块：

$$206.0.64.0/18::=\{<11001110.00000000.01>,<000000.00000000>\}$$

现在要将网络前缀后移四位，由于CIDR中也并不支持全0或者全1的主机号或网络号，因此分配给大学的地址块可以有$14(2^{4}-2)$种选项：

$$206.0.68.0/22::=\{<11001110.00000000.010001>,<00.00000000>\}$$
$$206.0.72.0/22::=\{<11001110.00000000.010010>,<00.00000000>\}$$
$$206.0.76.0/22::=\{<11001110.00000000.010011>,<00.00000000>\}$$
$$...$$
$$206.0.120.0/22::=\{<11001110.00000000.011110>,<00.00000000>\}$$

现在就选取第一种地址块的划分方案：$206.0.68.0/22::=\{<11001110.00000000.010001>,<00.00000000>\}$。现在大学又要在四个系中分配网络，已知四个系中主机需求最少的只要求支持100台即可，因此需要有$7(2^{7}=128)$位来代表主机，故网络前缀数目为$25(32-7)$位，那么就有如下$8(2^{25-22})$种划分方案：

$$206.0.68.0/25::=\{<11001110.00000000.01000100.0>,<0000000>\}$$
$$206.0.68.128/25::=\{<11001110.00000000.01000100.1>,<0000000>\}$$
$$206.0.69.0/25::=\{<11001110.00000000.01000101.0>,<0000000>\}$$
$$206.0.69.128/25::=\{<11001110.00000000.01000101.1>,<0000000>\}$$
$$206.0.70.0/25::=\{<11001110.00000000.01000110.0>,<0000000>\}$$
$$206.0.70.128/25::=\{<11001110.00000000.01000110.1>,<0000000>\}$$
$$206.0.71.0/25::=\{<11001110.00000000.01000111.0>,<0000000>\}$$
$$206.0.71.128/25::=\{<11001110.00000000.01000111.1>,<0000000>\}$$

可以将最后两个地址块分配给主机需求为100的两个系，即$206.0.71.0/25$分配给3系，$206.0.71.128/25$分配给四系。剩下的地址可以聚合成超网来满足更大规模的主机需求。2系有200个主机需求，因此需要的主机号位数为$8(2^{8}=256)$，可以考虑要么分配给2系两个主机容量为128的地址块，要么将两个主机容量为128的地址块合并为1个，第一种方案就已经不需要我们做什么了，第二种方案我们需要找两个连续的地址块来合并：在这里我们可以选择$206.0.70.0/25$和$206.0.70.128/25$来合并，这是因为他们是连续的而且其合并的那一位中0和1都出现过了，合并结果如下：

$$206.0.70.0/24::=\{<11001110.00000000.01000110>.<00000000>\}$$

最后一个系需要一个能容纳400个主机的地址块，那么就只能合并最上面四个地址块，刚好上面四个地址块是连续的也把要合并的2位的所有情况都包含了，合并后的结果如下：

$$206.0.68.0/23::=\{<11001110.00000000.0100010>,<0.00000000>\}$$

最后的结果总结为：

|组织/单位|地址块|二进制表示|地址数/容纳的主机数|
|:-:|:-:|:-:|:-:|
|ISP|206.0.64.0/18|<11001110.00000000.01>,<000000.00000000>|16384|
|大学|206.0.68.0/22|<11001110.00000000.010001>,<00.00000000>|1024|
|一系|206.0.68.0/23|<11001110.00000000.0100010>,<0.00000000>|512|
|二系|206.0.70.0/24|<11001110.00000000.01000110>，<00000000>|256|
|三系|206.0.71.0/25|<11001110.00000000.01000111.0>,<0000000>|128|
|四系|206.0.71.128/25|<11001110.00000000.01000111.1>,<0000000>|128|

四个系得到地址块以后可以再次进行各自的分配。
